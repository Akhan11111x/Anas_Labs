```python
import tensorflow as tf

from tensorflow.keras.preprocessing.image import ImageDataGenerator

import matplotlib.pyplot as plt

import os


print("Libraries imported successfully")
```

```python
dataset_path = "image_dataset"


for folder in os.listdir(dataset_path):
    
    print(folder)
```

```python
augmentation = ImageDataGenerator(
    
    rotation_range = 40,
    
    width_shift_range = 0.2,
    
    height_shift_range = 0.2,
    
    zoom_range = 0.3,
    
    horizontal_flip = True
)


print("Image augmentation created successfully")
```

```python
from tensorflow.keras.preprocessing import image
import numpy as np


img_path = "image_dataset/cats/" + os.listdir("image_dataset/cats")[0]


img = image.load_img(
    img_path,
    target_size=(150,150)
)


img_array = image.img_to_array(img)


img_array = np.expand_dims(
    img_array,
    axis=0
)


print("Image loaded successfully")
```

```python
i = 0

plt.figure(figsize=(10,5))


for batch in augmentation.flow(
    img_array,
    batch_size=1
):

    plt.subplot(1,5,i+1)

    plt.imshow(
        batch[0].astype("uint8")
    )

    plt.axis("off")

    i = i + 1

    if i == 5:
        break


plt.show()
```

```python
save_path = "augmented_images"


os.makedirs(
    save_path,
    exist_ok=True
)


i = 0

for batch in augmentation.flow(
    img_array,
    batch_size=1,
    save_to_dir=save_path,
    save_prefix="augmented",
    save_format="jpg"
):

    i = i + 1

    if i == 10:
        break


print("Augmented images saved successfully")
```

## Conclusion

In this lab, image data augmentation was performed using Keras ImageDataGenerator.

Different augmentation techniques including rotation, shifting, zooming, and flipping were applied.

New variations of images were generated from the original dataset.

Image augmentation helps increase dataset diversity and improves deep learning model performance.
