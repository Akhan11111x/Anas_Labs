```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image

import matplotlib.pyplot as plt
import numpy as np
import os

print("Libraries imported successfully")
```

```python
dataset_path = "image_dataset"

for folder in os.listdir(dataset_path):
    print(folder)
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.3,
    shear_range=0.2,
    horizontal_flip=True,
    fill_mode="nearest"
)

print("ImageDataGenerator created successfully")
```

```python
img_path = "image_dataset/cats/" + os.listdir("image_dataset/cats")[0]

img = image.load_img(
    img_path,
    target_size=(150,150)
)

img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)

print("Image loaded successfully")
```

```python
plt.figure(figsize=(12,5))

i = 0

for batch in datagen.flow(img_array, batch_size=1):
    plt.subplot(1,5,i+1)
    plt.imshow(batch[0].astype("uint8"))
    plt.axis("off")

    i = i + 1

    if i == 5:
        break

plt.show()
```

```python
save_path = "augmented_images_lab9"

os.makedirs(save_path, exist_ok=True)

i = 0

for batch in datagen.flow(
    img_array,
    batch_size=1,
    save_to_dir=save_path,
    save_prefix="augmented",
    save_format="jpg"
):
    i = i + 1

    if i == 10:
        break

print("Augmented images saved successfully")
```

## Conclusion

In this lab, Keras ImageDataGenerator was used for image augmentation.

Different transformations such as rotation, shifting, zooming, shearing, and flipping were applied.

Augmented images were displayed and saved successfully.

This technique helps increase dataset variety and improves deep learning model performance.
