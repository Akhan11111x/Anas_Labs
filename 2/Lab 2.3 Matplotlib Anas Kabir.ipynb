```python
import matplotlib.pyplot as plt
import numpy as np


print("Libraries imported successfully")
```

```python
x = np.array([1,2,3,4,5])

y = np.array([2,4,6,8,10])


plt.plot(x,y)


plt.show()
```

```python
plt.plot(x,y)


plt.title("Simple Line Graph")


plt.xlabel("X Values")


plt.ylabel("Y Values")


plt.show()
```

```python
x = np.array([1,2,3,4,5])


y1 = np.array([2,4,6,8,10])

y2 = np.array([1,3,5,7,9])


plt.plot(x,y1,label="Line 1")


plt.plot(x,y2,label="Line 2")


plt.title("Multiple Line Graph")


plt.xlabel("X Values")


plt.ylabel("Y Values")


plt.legend()


plt.show()
```

```python
students = ["Ali","Ahmed","Sara","Anas"]

marks = [85,90,78,92]


plt.bar(students,marks)


plt.title("Student Marks Bar Chart")


plt.xlabel("Students")


plt.ylabel("Marks")


plt.show()
```

```python
hours = [1,2,3,4,5]

scores = [40,50,65,75,90]


plt.scatter(hours,scores)


plt.title("Study Hours vs Scores")


plt.xlabel("Study Hours")


plt.ylabel("Scores")


plt.show()
```

```python
subjects = ["Math", "Science", "English", "Computer"]

marks = [30,25,20,25]


plt.pie(
    marks,
    labels=subjects,
    autopct="%1.1f%%"
)


plt.title("Subject Marks Distribution")


plt.show()
```

```python
x = [1,2,3,4,5]

y = [10,20,30,40,50]


plt.plot(x,y)


plt.title("Saved Line Graph")


plt.xlabel("X Data")


plt.ylabel("Y Data")


plt.savefig("matplotlib_graph.png")


plt.show()


print("Graph saved successfully")
```
