```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('5_features_extracted.csv')

df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[['aep']] = scaler.fit_transform(df[['aep']])

df.head()
```

```python
df = pd.get_dummies(
    df,
    columns=['month', 'day_of_week'],
    dtype=int
)

df.head()
```

```python
df['hour_sin'] = np.sin(2 * np.pi * df['hour']/24)

df['hour_cos'] = np.cos(2 * np.pi * df['hour']/24)

df.head()
```

```python
df.to_csv('6_AEP_normalized_encoded.csv', index=False)

print("Normalized and encoded CSV saved successfully")
```
