```python
import numpy as np

print("Numpy imported successfully")
```

```python
arr = np.array([1, 2, 3, 4, 5])

print(arr)
```

```python
print(type(arr))
```

```python
arr2 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(arr2)
```

```python
print("Dimension of Array:")

print(arr2.ndim)


print("Shape of Array:")

print(arr2.shape)


print("Size of Array:")

print(arr2.size)
```

```python
numbers = np.arange(1, 11)


print(numbers)
```

```python
zeros_array = np.zeros((3,3))

ones_array = np.ones((3,3))


print("Zeros Array:")

print(zeros_array)


print("Ones Array:")

print(ones_array)
```

```python
random_array = np.random.rand(3,3)


print("Random Array:")

print(random_array)
```

```python
array = np.arange(1,13)


reshape_array = array.reshape(3,4)


print("Original Array:")

print(array)


print("Reshaped Array:")

print(reshape_array)
```

```python
print("First Element:")

print(arr[0])


print("Last Element:")

print(arr[-1])


print("2D Array Element:")

print(arr2[1,2])
```

```python
array1 = np.array([10,20,30,40])

array2 = np.array([1,2,3,4])


print("Addition:")

print(array1 + array2)


print("Subtraction:")

print(array1 - array2)


print("Multiplication:")

print(array1 * array2)


print("Division:")

print(array1 / array2)
```

```python
data = np.array([5,10,15,20,25])


print("Mean:")

print(np.mean(data))


print("Maximum Value:")

print(np.max(data))


print("Minimum Value:")

print(np.min(data))


print("Standard Deviation:")

print(np.std(data))
```

```python
np.save(
    "numpy_array.npy",
    data
)


print("Numpy array saved successfully")
```
