```python
import numpy as np

import matplotlib.pyplot as plt


print("Libraries imported successfully")
```

```python
current = np.linspace(
    1,
    10,
    50
)


resistance = 5


voltage = current * resistance


print("Ohm Law dataset created successfully")
```

```python
noise = np.random.normal(
    0,
    2,
    voltage.shape
)


voltage_noisy = voltage + noise


print("Noise added successfully")
```

```python
plt.scatter(
    current,
    voltage_noisy
)


plt.title("Ohm's Law Data")


plt.xlabel("Current")


plt.ylabel("Voltage")


plt.show()
```

```python
I_mean = current.mean()

I_std = current.std()


V_mean = voltage_noisy.mean()

V_std = voltage_noisy.std()


I_norm = (current - I_mean) / I_std

V_norm = (voltage_noisy - V_mean) / V_std


print("Data normalized successfully")
```

```python
m = 0

b = 0


learning_rate = 0.01

epochs = 1000


loss_history = []


print("Initial values created")
```

```python
n = len(I_norm)


for i in range(epochs):


    prediction = m * I_norm + b


    error = prediction - V_norm


    loss = (1/n) * np.sum(error**2)


    loss_history.append(loss)


    dm = (2/n) * np.sum(error * I_norm)

    db = (2/n) * np.sum(error)


    m = m - learning_rate * dm

    b = b - learning_rate * db


print("Gradient Descent Completed")


print("Slope:", m)

print("Intercept:", b)
```

```python
predicted_voltage = m * I_norm + b


plt.scatter(
    I_norm,
    V_norm,
    label="Actual Data"
)


plt.plot(
    I_norm,
    predicted_voltage,
    label="Regression Line"
)


plt.title("Ohm's Law Linear Regression")


plt.xlabel("Current")


plt.ylabel("Voltage")


plt.legend()


plt.show()
```

```python
plt.plot(loss_history)


plt.title("Gradient Descent Loss Curve")


plt.xlabel("Epochs")


plt.ylabel("Loss")


plt.show()
```
