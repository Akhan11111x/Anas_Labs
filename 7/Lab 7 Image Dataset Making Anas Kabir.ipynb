Objective:
To create an image dataset folder structure for machine learning classification tasks.

```python
import os
from PIL import Image
import matplotlib.pyplot as plt
```

```python
dataset_path = "image_dataset"

classes = ["cats", "dogs", "cars"]

for class_name in classes:
    folder_path = os.path.join(dataset_path, class_name)
    os.makedirs(folder_path, exist_ok=True)

print("Image dataset folders created successfully")
```

```python
for folder in os.listdir(dataset_path):
    print(folder)
```

```python
for folder in os.listdir(dataset_path):
    
    folder_location = os.path.join(dataset_path, folder)
    
    total_images = len(os.listdir(folder_location))
    
    print(folder, ":", total_images, "images")
```

```python
import matplotlib.image as mpimg

plt.figure(figsize=(10,5))

count = 1

for folder in os.listdir(dataset_path):

    folder_location = os.path.join(dataset_path, folder)

    image_name = os.listdir(folder_location)[0]

    image_path = os.path.join(folder_location, image_name)

    img = mpimg.imread(image_path)

    plt.subplot(1,3,count)

    plt.imshow(img)

    plt.title(folder)

    plt.axis("off")

    count = count + 1


plt.show()
```

## Conclusion
