Objective:
To understand underfitting, good fitting, and overfitting using polynomial regression on projectile motion data.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(42)

x = np.linspace(0, 10, 20)

y_true = -0.5 * x**2 + 5 * x + 2

noise = np.random.normal(0, 2, size=x.shape)

y = y_true + noise

print("Projectile motion data created successfully")
```

```python
plt.figure(figsize=(8,5))

plt.scatter(x, y, label="Noisy Projectile Data")

plt.xlabel("Horizontal Distance")
plt.ylabel("Height")

plt.title("Projectile Motion Data")

plt.legend()
plt.show()
```

```python
def polynomial_model(degree):

    poly = PolynomialFeatures(degree=degree)

    x_poly = poly.fit_transform(x.reshape(-1,1))

    model = LinearRegression()

    model.fit(x_poly, y)

    prediction = model.predict(x_poly)

    mse = mean_squared_error(y, prediction)

    return prediction, mse
```

```python
underfit, mse1 = polynomial_model(1)

goodfit, mse2 = polynomial_model(2)

overfit, mse3 = polynomial_model(15)


print("Underfitting Error:", mse1)

print("Good Fitting Error:", mse2)

print("Overfitting Error:", mse3)
```

```python
plt.figure(figsize=(10,6))

plt.scatter(x, y, label="Original Data")

plt.plot(x, underfit, label="Underfitting Degree 1")

plt.plot(x, goodfit, label="Good Fit Degree 2")

plt.plot(x, overfit, label="Overfitting Degree 15")

plt.xlabel("Horizontal Distance")
plt.ylabel("Height")

plt.title("Underfitting vs Good Fitting vs Overfitting")

plt.legend()

plt.show()
```

## Conclusion

In this lab, underfitting and overfitting concepts were studied using projectile motion data.

Polynomial regression models with different degrees were trained.

Degree 1 model showed underfitting because it was too simple.

Degree 2 model provided the best fit because projectile motion follows a quadratic relationship.

Degree 15 model showed overfitting because it captured unnecessary noise in the data.

This experiment shows the importance of selecting the correct model complexity in machine learning.
