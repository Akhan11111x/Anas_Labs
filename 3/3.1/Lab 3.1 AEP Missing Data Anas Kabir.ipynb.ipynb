Objective:
To identify and handle missing values in AEP energy dataset using Pandas preprocessing techniques.

```python
import pandas as pd

import numpy as np


print("Libraries imported successfully")
```

```python
data = {
    "Datetime": [
        "2024-01-01",
        "2024-01-02",
        "2024-01-03",
        "2024-01-04",
        "2024-01-05"
    ],

    "AEP_Load": [
        12300,
        12550,
        np.nan,
        11900,
        12150
    ],

    "Temperature": [
        25,
        np.nan,
        28,
        24,
        26
    ]
}


df = pd.DataFrame(data)


df
```

```python
print("Missing Values Count:")


print(df.isnull().sum())
```

```python
df["AEP_Load"] = df["AEP_Load"].fillna(
    df["AEP_Load"].mean()
)


df["Temperature"] = df["Temperature"].fillna(
    df["Temperature"].mean()
)


print("Missing values filled successfully")


df
```

```python
print("Missing Values After Cleaning:")


print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt


plt.plot(
    df["Datetime"],
    df["AEP_Load"]
)


plt.title("AEP Load After Missing Value Handling")


plt.xlabel("Date")


plt.ylabel("AEP Load")


plt.xticks(rotation=45)


plt.show()
```

```python
df.to_csv(
    "AEP_cleaned_data.csv",
    index=False
)


print("Cleaned AEP dataset saved successfully")
```
