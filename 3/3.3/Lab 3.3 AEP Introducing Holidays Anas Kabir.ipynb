```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt


print("Libraries imported successfully")
```

```python
data = {
    "Datetime": pd.date_range(
        start="2024-01-01",
        periods=10,
        freq="D"
    ),

    "AEP_Load":[
        12300,
        12500,
        12100,
        11950,
        12400,
        12600,
        11800,
        12250,
        12700,
        12000
    ]
}


df = pd.DataFrame(data)


df
```

```python
df["Year"] = df["Datetime"].dt.year


df["Month"] = df["Datetime"].dt.month


df["Day"] = df["Datetime"].dt.day


df
```

```python
holiday_dates = [
    "2024-01-01",
    "2024-01-05"
]


df["Holiday"] = df["Datetime"].isin(
    pd.to_datetime(holiday_dates)
)


df
```

```python
df["Holiday"] = df["Holiday"].astype(int)


print("Holiday converted successfully")


df
```

```python
holiday_compare = df.groupby(
    "Holiday"
)["AEP_Load"].mean()


print(holiday_compare)
```

```python
holiday_compare.plot(
    kind="bar"
)


plt.title("Holiday Effect on AEP Load")


plt.xlabel("Holiday Status")


plt.ylabel("Average Load")


plt.show()
```

```python
df.to_csv(
    "AEP_holiday_dataset.csv",
    index=False
)


print("AEP holiday dataset saved successfully")
```
