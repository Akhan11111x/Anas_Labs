Objective:
To design and train a 1D Convolutional Neural Network model for sequence classification.

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense

print("Libraries imported successfully")
```

```python
X = np.array([
    [1,2,3,4,5,6],
    [2,3,4,5,6,7],
    [10,11,12,13,14,15],
    [11,12,13,14,15,16],
    [3,4,5,6,7,8],
    [12,13,14,15,16,17]
])

Y = np.array([0,0,1,1,0,1])

print("1D dataset created successfully")
```

```python
X = X.reshape(
    X.shape[0],
    X.shape[1],
    1
)

print("Data reshaped successfully")
print("Shape of X:", X.shape)
```

```python
model = Sequential()


model.add(Conv1D(
    filters=32,
    kernel_size=2,
    activation="relu",
    input_shape=(6,1)
))


model.add(MaxPooling1D(
    pool_size=2
))


model.add(Flatten())


model.add(Dense(
    16,
    activation="relu"
))


model.add(Dense(
    1,
    activation="sigmoid"
))


print("1D CNN model created successfully")
```

```python
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)


print("1D CNN model compiled successfully")
```

```python
history = model.fit(
    X,
    Y,
    epochs=50,
    verbose=1
)


print("1D CNN training completed successfully")
```

```python
plt.figure(figsize=(8,5))

plt.plot(history.history['accuracy'])

plt.xlabel("Epochs")

plt.ylabel("Accuracy")

plt.title("1D CNN Training Accuracy")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(history.history['loss'])

plt.xlabel("Epochs")

plt.ylabel("Loss")

plt.title("1D CNN Training Loss")

plt.show()
```

```python
prediction = model.predict(X)


print("Predicted Values:")

print(prediction)


print("Actual Values:")

print(Y)
```

```python
model.summary()
```

```python
model.save("1d_cnn_model.h5")


print("1D CNN model saved successfully")
```
