```python
import pandas as pd

print("Pandas imported successfully")
```

```python
marks = pd.Series([85, 90, 78, 92, 88])

print(marks)
```

```python
data = {
    "Name": ["Ali", "Ahmed", "Sara", "Anas", "Zara"],
    "Age": [20, 21, 19, 22, 20],
    "Marks": [85, 90, 78, 92, 88]
}

df = pd.DataFrame(data)

df
```

```python
print("First five rows:")

print(df.head())


print("Data Information:")

df.info()
```

```python
df.describe()
```

```python
names = df["Name"]

marks = df["Marks"]


print("Student Names:")

print(names)


print("Student Marks:")

print(marks)
```

```python
df["Grade"] = [
    "B",
    "A",
    "C",
    "A",
    "B"
]


df
```

```python
high_marks = df[df["Marks"] >= 85]


print("Students with high marks:")


high_marks
```

```python
sorted_data = df.sort_values(
    by="Marks",
    ascending=False
)


sorted_data
```

```python
df.to_csv(
    "students_data.csv",
    index=False
)


print("CSV file saved successfully")
```

```python
new_data = pd.read_csv(
    "students_data.csv"
)


new_data
```

```python
print("Missing Values:")


print(new_data.isnull().sum())
```
