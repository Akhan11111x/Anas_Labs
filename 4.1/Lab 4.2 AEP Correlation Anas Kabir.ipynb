Objective: To check correlation between AEP electricity load and extracted features.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('5_features_extracted.csv')

df.head()
```

```python
df.info()
```

```python
correlation = df.corr(numeric_only=True)

correlation
```

```python
correlation['aep'].sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,6))

plt.bar(
    correlation['aep'].index,
    correlation['aep'].values
)

plt.xticks(rotation=90)

plt.xlabel("Features")
plt.ylabel("Correlation Value")

plt.title("Correlation of Features with AEP Load")

plt.show()
```

## Conclusion

In this lab, correlation analysis was performed on AEP energy dataset.

Different time-based features were compared with AEP load values.

Correlation helps to identify which features have stronger relationship with electricity consumption and which features are less important for prediction models.
