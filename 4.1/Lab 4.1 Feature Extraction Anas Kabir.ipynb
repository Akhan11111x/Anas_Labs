```python
import pandas as pd

data = {
    'Datetime': [
        '2024-01-01 01:00:00',
        '2024-01-01 02:00:00',
        '2024-03-15 03:00:00',
        '2024-07-20 04:00:00',
        '2024-10-10 05:00:00'
    ],
    'AEP_MW': [
        12379,
        11935,
        11692,
        11597,
        11681
    ],
    'Date': [
        '2024-01-01',
        '2024-01-01',
        '2024-03-15',
        '2024-07-20',
        '2024-10-10'
    ],
    'Holiday': [
        0,0,0,0,0
    ]
}

df = pd.DataFrame(data)

df.to_csv('4_AEP_Introducing_holidays.csv',
          index=False)

print("CSV file created successfully")
```

```python
df = pd.read_csv('4_AEP_Introducing_holidays.csv',
                 parse_dates=['Datetime'])

df.head()
```

```python
df['Hour'] = df['Datetime'].dt.hour

df['Month'] = df['Datetime'].dt.month

df['Day_Of_Week'] = df['Datetime'].dt.weekday

df['Week'] = df['Datetime'].dt.isocalendar().week

df['yearday'] = df['Datetime'].dt.dayofyear

df['quarter'] = df['Datetime'].dt.quarter

df['weekend'] = (df['Day_Of_Week'] >= 5).astype("int")

df['day_night'] = ((df['Hour'] >= 8) & (df['Hour'] <= 16)).astype("int")

df.head()
```

```python
df['winter'] = ((df['Month'] == 12) | 
                (df['Month'] == 1) | 
                (df['Month'] == 2)).astype("int")

df['spring'] = ((df['Month'] == 3) | 
                (df['Month'] == 4) | 
                (df['Month'] == 5)).astype("int")

df['summer'] = ((df['Month'] == 6) | 
                (df['Month'] == 7) | 
                (df['Month'] == 8)).astype("int")

df['fall'] = ((df['Month'] == 9) | 
              (df['Month'] == 10) | 
              (df['Month'] == 11)).astype("int")

df.tail()
```

```python
df.rename(columns={
    'AEP_MW': 'aep',
    'yearday': 'year_day',
    'Week': 'week_no',
    'Holiday': 'holiday',
    'Hour': 'hour',
    'Month': 'month',
    'Day_Of_Week': 'day_of_week'
}, inplace=True)

df.head()
```

```python
df = df.reindex(columns=[
    'Datetime', 'aep', 'year_day', 'holiday', 'weekend',
    'winter', 'spring', 'summer', 'fall',
    'hour', 'month', 'day_of_week'
])

df.head()
```

```python
df.to_csv('5_features_extracted.csv', index=False)

print("Feature extracted CSV saved successfully")
```
